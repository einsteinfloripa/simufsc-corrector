```python
from itertools import combinations
# ele pega as possiveis alternativas e faz a permutação entre elas
# dessa forma, temos todas as possiveis somas que os alunos podem responder e as alternativas que ele assinalou em tal soma

alternativas = [1, 2 , 4, 8, 16, 32, 64]

d = {} 

j = 0
while j < len(alternativas):
    comb = combinations(alternativas, j)
    
    for i in list(comb):
        d.update({
            ','.join(map(str, (i))) : int(sum(list(i)))
        })
    j += 1
```

```python
# aqui ele pega as combinações e cria um Dataframe com as possíveis somas, e as alternativas que forma assinaladas pra calcular a soma
possiveis_somas = pd.DataFrame(list(d.items()), columns = ['alternativas','soma']) 
possiveis_somas.head(10)
```

## 3.1 Destrinchando as somatorias

As funções a seguir lêem o valor da célula de resposta do aluno e comparam com os valores do df 'possiveis_somas'

```python
def somatoria_resp(row):
    soma = 0
    if (row["NP"] != "ABERTA") and (row["resp_aluno"] != 0): 
        soma = int(row["resp_aluno1"])
        assinaladas = possiveis_somas['alternativas'][soma == possiveis_somas['soma']]
        return str(assinaladas.values)
    else:
        return str(row["resp_aluno"])
```

```python
def somatoria_gab(row):
    if (row["NP"] != "ABERTA"): 
        soma = row["Gabarito"]
        assinaladas = possiveis_somas['alternativas'][soma == possiveis_somas['soma']]
        return str(assinaladas.values)
    else:
        return str(row["Gabarito"])
```

As funções a seguir criam 14 colunas, uma coluna para cada alternativa (01,02,04,08,16,32,64) para a resposta do aluno e para o gabarito. 

Se o aluno assinalou a alternativa X, a coluna "resp_X" vai ficar com valor 1, caso contrário, 0.

```python
def is_or_not_resp(row):
    assinaladas = row['assinaladas']
    assinaladas = re.findall(r"[\w']+", assinaladas)

    for j in list(alternativas):
        for i in assinaladas:
            if row['assinaladas'] != "" and row['assinaladas'] != "None" and str(j) == i:
                
                row["resp_" + str(j)] = 1
    return row
```

```python
def is_or_not_gab(row):
    certas = row['alternativas_certas']
    certas = re.findall(r"[\w']+", certas)

    for j in list(alternativas):
        for i in certas:
            if row['alternativas_certas'] != "" and row['alternativas_certas'] != "None" and str(j) == i:
                row["gab_" + str(i)] = 1
    return row
```

```python
df = df.apply(is_or_not_resp, axis=1)
```

```python
# calculo da pontuação
# a função vai linha por linha do df e analisa cada coluna de cada alternativa 
# se o dado da coluna resposta do aluno (resp_XX) é igual ao dado do gabarito (gab_XX)
# e a resposta estiver correta (gab_XX == 1), soma-se 1 ao numero de proposições corretas (NPC)

# se a resposta for diferente do gabarito e o aluno assinalou essa alternativa (resp_XX == 1)
# soma-se 1 ao numero de proposições incorretas(NPI)

def correction(row):
    npc = 0
    npi = 0
    ntpc = 0
    i = 0
    n = 0
    
    # se a quesão não é aberta, tem acerto parcial. Caso a questão sea aberta, não tem parcial
    if (row["NP"] != "ABERTA"):
        while n < int(row['NP']):

            numero = alternativas[i]
            ntpc += int(row["gab_" + str(numero)])

            if row["gab_" + str(numero)] == row["resp_" + str(numero)] and row["resp_" + str(numero)] == 1.0:
                npc += 1
                
            elif row["gab_" + str(numero)] != row["resp_" + str(numero)] and row["resp_" + str(numero)] == 1.0:
                npi += 1
                
            n += 1
            i += 1
            
        # pontuacao
        if (npc > npi):
            p = (int(row['NP']) - (ntpc -(npc - npi))) / int(row['NP'])
        else:
            p = 0
    else:
        if row["resp_aluno"] == row["gabarito"]:
            p = 1
        else:
            p = 0
    return round(p, 2)
```

```python
df['pontuacao'] = df.apply(correction, axis=1)
```
