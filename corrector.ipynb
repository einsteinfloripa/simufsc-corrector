## Destrinchando as somatorias

As funções a seguir lêem o valor da célula de resposta do aluno e comparam com os valores do df 'possiveis_somas'
Depois de comparar, ele pega as alternativas assinaladas e insere uma nova coluna no 'df'

```python
# calculo da pontuação
# a função vai linha por linha do df e analisa se o dado da resposta do aluno é igual ao dado do gabarito
# se a resposta for igual ao gabarito e a resposta estiver correta, soma-se 1 ao numero de proposições corretas (NPC)
# se a resposta for diferente do gabarito e a o aluno assinalou essa alternativa, soma-se 1 ao numero de proposições incorretas (NPI)

def correction(row):
    if (row["NP"] != "ABERTA"):
        npc = 0
        npi = 0
        ntpc = 0
        i = 0
        n = 0
        while n < row['NP']:

            numero = alternativas[i]
            ntpc += int(row["gab_" + str(numero)])

            if row["gab_" + str(numero)] == row["resp_" + str(numero)] & row["resp_" + str(numero)] == 1:
                npc += 1
                
            elif row["gab_" + str(numero)] != row["resp_" + str(numero)] & row["resp_" + str(numero)] == 1:
                npi += 1
                
            n += 1
            i += 1

        if (npc > npi):
            p = (row['NP'] - ntpc + npc - npi) / row['NP']
        else:
            p = 0
    else:
        if row["resp_aluno"] == row["gabarito"]:
            p = 1
        else:
            p = 0
    return round(p, 2)
```

```python
df['pontuacao'] = df.apply(correction, axis=1)
```
