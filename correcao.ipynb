### Somatórias

```python
from itertools import combinations

alternativas = [1, 2 , 4, 8, 16, 32, 64]

d = {} 

j = 0
while j < len(alternativas):
    comb = combinations(alternativas, j)
    
    for i in list(comb):
        d.update({
            ';'.join(map(str, (i))) : int(sum(list(i)))
        })
    j += 1
```

```python
possiveis_somas = pd.DataFrame(list(d.items()), columns = ['alternativas','soma']) 
possiveis_somas
```

## Manipulação

```python
def somatoria_resp(row):
    soma = row["resp_aluno"]
    assinaladas = possiveis_somas['alternativas'][soma == possiveis_somas['soma']]
    return str(assinaladas.values)
```

```python
def somatoria_gab(row):
    soma = row["gabarito"]
    assinaladas = possiveis_somas['alternativas'][soma == possiveis_somas['soma']]
    return str(assinaladas.values)
```

```python
df['assinaladas'] = df.apply(somatoria_resp, axis=1).astype(str)
df['alternativas_certas'] = df.apply(somatoria_gab, axis=1).astype(str)
```

```python
df['assinaladas'] = df['assinaladas'].str.strip('[]') #.apply(lambda x: x.replace('[','').replace(']',''))
df['alternativas_certas'] = df['alternativas_certas'].str.strip('[]')
```

```python
def correction(row):
    
    npc = 0
    npi = 0
    ntpc = 0
    i = 0
    n = 0
    while n < row['np']:
        
        numero = alternativas[i]
        ntpc += int(row["gab_" + str(numero)])
        
        if row["gab_" + str(numero)] == row["resp_" + str(numero)] & row["resp_" + str(numero)] == 1:
            npc += 1
            # print("acertou" + str(row["gab_" + str(i)]) + str(row["resp_" + str(i)]))
        # df["gab_" + str(i)] = df['alternativas_certas'].apply(lambda x: 1 if str(i) in x else 0)
        elif row["gab_" + str(numero)] != row["resp_" + str(numero)] & row["resp_" + str(numero)] == 1:
            npi += 1
        n += 1
        i += 1
        
    
    if (npc > npi):
        p = (row['np'] - ntpc + npc - npi) / row['np']
    else:
        p = 0
    
    return round(p, 2)
```

```python
df['pontuacao'] = df.apply(correction, axis=1)
```
